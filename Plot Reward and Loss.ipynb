```python
import pandas as pd

df = pd.read_csv('vanilla_dqn_boltzmann_cartpole_1000_Episode/info/vanilla_dqn_boltzmann_cartpole_1000_t0_s0_session_df_train.csv')
```

```python
import matplotlib.pyplot as plt

plt.plot(df['epi'], df['total_reward'])
plt.title('Total During over 1000 Episodes')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.ylim(5,15)
plt.show()
```
